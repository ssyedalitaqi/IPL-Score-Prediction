```python
from pyforest import *
```

```python
data=pd.read_csv('ipl.csv')
```

```python
data
```

```python
data=data[['date','venue','bat_team','bowl_team','runs','overs','runs_last_5','wickets_last_5','total']]
```

```python
data.head()
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data.dtypes
```

```python
data['year']=data['date'].dt.year
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data['bat_team'].unique()
```

```python
data['bowl_team'].unique()
```

```python
data['bowl_team'].unique()==data['bat_team'].unique()
```

```python
data.drop('date',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['bat_team']=data['bat_team'].replace('Rising Pune Supergiants','Rising Pune Supergiant')
data['bowl_team']=data['bowl_team'].replace('Rising Pune Supergiants','Rising Pune Supergiant')
```

```python
len(data['bat_team'].unique())
```

```python
data=data.loc[data['overs']>=5.0,:]
```

```python
data=pd.get_dummies(data=data,columns=['bat_team','bowl_team'])
```

```python
x_train=data.loc[data['year']<=2016,:]
x_test=data.loc[data['year']>=2016,:]
```

```python
y_train=data.loc[data['year']<=2016,['total']]
y_test=data.loc[data['year']>=2016,['total']]
```

```python
x_train.drop(['total','year'],axis=1,inplace=True)
x_test.drop(['total','year'],axis=1,inplace=True)
```

```python
x_train
```

```python
y_test
```

```python
x_train.drop('venue',axis=1,inplace=True)
x_test.drop('venue',axis=1,inplace=True)
```

```python
x_test.head()
```

```python
x_train.head()
```

**Model Building**

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
lr_model.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
svm_model=SVR()
svm_model.fit(x_train,y_train)
svm_model.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model=KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train,y_train)
knn_model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr_model=RandomForestRegressor()
rfr_model.fit(x_train,y_train)
rfr_model.score(x_test,y_test)
```

```python
import pickle
pickle.dump(rfr_model,open('rfr_model.pkl','wb'))
```
